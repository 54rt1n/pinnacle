```python
# This cell is used for imports and setup

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import noise
from worldgen.location import Location

config = {
    'size': (128, 128),
    'seed': 1250,
    'noise_scale': 100,
    'num_biomes': 5,
    'temperature': 20,  # Celsius
    'humidity': 0.6     # Relative humidity
}

location = Location.from_config(config)
res = location.generate()
```

```python
# This cell is used for visualization

fig, ax = plt.subplots(2, 1, figsize=(10, 20))

# Visualize the heightmap
ax[0].imshow(res['heightmap'], cmap='terrain')
ax[0].set_title('Heightmap')

# Visualize the biomes
ax[1].imshow(res['biomes'], cmap='tab10')
ax[1].set_title('Biomes')

plt.show()
```
